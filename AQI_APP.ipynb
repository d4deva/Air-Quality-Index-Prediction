```python
from tkinter import *
from PIL import ImageTk
import pickle
model = pickle.load(open("AQI_RFR_model.pkl","rb"))
class AQI:
    def __init__(self,report):
        self.root = root
        root.iconbitmap("icon_aqi.ico")
        self.root.title("Air Quality Index")
        self.root.geometry("1199x600")
        self.root.resizable(False,False)
        #==== BG IMAGE ========
        self.bg = ImageTk.PhotoImage(file="Wor.jpg")
        self.bg_image = Label(self.root,image=self.bg).place(x=0,y=0,relwidth=1,relheight=1)
        ##### AQI Frames #######
        frame_AQI = Frame(self.root,bg="white")
        frame_AQI.place(x=600,y=50,height=500,width=500)
        
        
        
        title = Label(frame_AQI,text="Air Qauality Index",font=("Arial",25,"bold","underline"),fg="black",bg="white").place(x=125,y=25)
        title2 = Label(frame_AQI,text="Fill in the details to calculate AQI :",font=("Arial",15,"bold","underline"),fg="black",bg="white").place(x=35,y=75)
        
        T = Label(frame_AQI,text="Average Temperature",font=("Arial",12),fg="black",bg="white").place(x=35,y=110)
        TM = Label(frame_AQI,text="Maximum temperature",font=("Arial",12),fg="black",bg="white").place(x=35,y=140)
        Tm = Label(frame_AQI,text="Minimum temperature",font=("Arial",12),fg="black",bg="white").place(x=35,y=170)
        H = Label(frame_AQI,text="Average relative humidity",font=("Arial",12),fg="black",bg="white").place(x=35,y=200)
        PP = Label(frame_AQI,text="Total rainfall",font=("Arial",12),fg="black",bg="white").place(x=35,y=230)
        VV = Label(frame_AQI,text="Average visibility",font=("Arial",12),fg="black",bg="white").place(x=35,y=260)
        V = Label(frame_AQI,text="Average wind speed",font=("Arial",12),fg="black",bg="white").place(x=35,y=290)
        VM = Label(frame_AQI,text="Sustained wind speed",font=("Arial",12),fg="black",bg="white").place(x=35,y=320)
        
        
        self.Ttext = Entry(frame_AQI,bg="light grey")
        self.Ttext.grid(row=1, column=1)
        self.Ttext.place(x=225,y=110,height=25,width=150)
        
        self.TMtext = Entry(frame_AQI,bg="light grey")
        self.TMtext.grid(row=1, column=1)
        self.TMtext.place(x=225,y=140,height=25,width=150)
        
        self.Tmtext = Entry(frame_AQI,bg="light grey")
        self.Tmtext.grid(row=1, column=1)
        self.Tmtext.place(x=225,y=170,height=25,width=150)
        
        self.Htext = Entry(frame_AQI,bg="light grey")
        self.Htext.grid(row=1, column=1)
        self.Htext.place(x=225,y=200,height=25,width=150)
        
        
        self.PPtext = Entry(frame_AQI,bg="light grey")
        self.PPtext.grid(row=1, column=1)
        self.PPtext.place(x=225,y=230,height=25,width=150)
        
        self.VVtext = Entry(frame_AQI,bg="light grey")
        self.VVtext.grid(row=1, column=1)
        self.VVtext.place(x=225,y=260,height=25,width=150)
        
        
        
        self.Vtext = Entry(frame_AQI,bg="light grey")
        self.Vtext.grid(row=1, column=1)
        self.Vtext.place(x=225,y=290,height=25,width=150)
        
        
        self.VMtext = Entry(frame_AQI,bg="light grey")
        self.VMtext.grid(row=1, column=1)
        self.VMtext.place(x=225,y=320,height=25,width=150)
        
        
        def get_values():
            
            T_value = float(self.Ttext.get())
            TM_value = float(self.TMtext.get())
            Tm_value = float(self.Tmtext.get())
            H_value = float(self.Htext.get())
            PP_value = float(self.PPtext.get())
            VV_value = float(self.VVtext.get())
            V_value = float(self.Vtext.get())
            VM_value = float(self.VMtext.get())
            
            pred_values = [[T_value,TM_value,Tm_value,H_value,PP_value,VV_value,V_value,VM_value]]
            pred = model.predict(pred_values)
            pred = pred.round(2)
            if pred <=100:
                result = Label(frame_AQI,text="The Air Quality index is {}".format(pred),font=("Arial",15,"bold"),fg="black",bg="forest green").place(x=50,y=410)
            elif pred>100 and pred<=200:
                result = Label(frame_AQI,text="The Air Quality index is {}".format(pred),font=("Arial",15,"bold"),fg="black",bg="yellow").place(x=50,y=410)                
            else:
                result = Label(frame_AQI,text="The Air Quality index is {}".format(pred),font=("Arial",15,"bold"),fg="black",bg="red").place(x=50,y=410)

                
            
            
            
        button = Button(frame_AQI,text ="Predict",font=("Arial",12,"bold"),fg="black",bg="grey",command = get_values).place(x=185,y=365)
        
        
            
root=Tk()
obj=AQI(root)
root.mainloop()

```
